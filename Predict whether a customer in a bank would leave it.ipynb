```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("BankCustomers.csv")
```

```python
df
```

```python
X = df.iloc[:, 3:13]
X
```

```python
y = df.iloc[:, 13]
y
```

```python
X["Geography"].value_counts()
```

```python
# Convert categorical feature into dummy variables

states = pd.get_dummies(X["Geography"], drop_first=True)
gender = pd.get_dummies(X["Gender"], drop_first=True)
```

```python
# Concatenate the remaining dummies columns

X = pd.concat([X,states,gender], axis=1)
```

```python
# Drop unused columns

X = X.drop(["Geography", "Gender"], axis=1)
```

```python
X
```

```python
# Splitting the dataset into the training set and test set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
# Build teh Artificial Neural Network

import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Initializing the ANN
classifier = Sequential()

# Adding the input layer and the fisrt hidden layer
#for each neurone we have sum Ii*Wi 
classifier.add(Dense(activation="relu", input_dim=11, units=6, kernel_initializer="uniform"))

# Adding the second hidden layer
classifier.add(Dense(activation="relu", units=6, kernel_initializer="uniform"))

# Adding the output layer
classifier.add(Dense(activation="sigmoid", units=1, kernel_initializer="uniform"))
```

```python
classifier.summary()
```

```python
# Compiling the ANN
classifier.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
# Fitting the ANN to the training set
classifier.fit(X_train, y_train, batch_size=10, epochs=100)
```

```python
# Predicting the test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
# Making the confusion matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
```
